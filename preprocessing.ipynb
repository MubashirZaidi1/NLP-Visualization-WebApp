### UTTERENCE CLEANING

```python
# # Download stopwords and wordnet if not already downloaded
# nltk.download('stopwords')
# nltk.download('wordnet')
```

```python
def cleanUtterences(text):
    text = text.replace("'", "")
    words = nltk.word_tokenize(text)
    cleaned_words = []
    
    stop_words = set(stopwords.words('english'))
    
    for word in words:
        # Remove punctuation character by character
        cleaned_word = ''
        for char in word:
            if char not in string.punctuation:
                cleaned_word += char
        if cleaned_word and cleaned_word.lower() not in stop_words:
            cleaned_words.append(cleaned_word)
    
    return cleaned_words
```

```python
# Apply the preprocessing function to the 'meta.parsed' column and create a new column 'updated_utt'
uttDF['updated_utt'] = uttDF['text'].apply(cleanUtterences)
```

```python
def cleanMovieNames(text):
    text = text.replace("'", "")
    text = text.replace("`", "")
    text = text.replace("``", "")        
    words = nltk.word_tokenize(text)
    cleaned_words = []
    stop_words = set(stopwords.words('english'))
    
    for word in words:
        cleaned_word = ''
        for char in word:
            if char not in string.punctuation:
                cleaned_word += char
        
        if cleaned_word and cleaned_word not in stop_words:
            cleaned_words.append(cleaned_word)
    
    cleaned_text = ' '.join(cleaned_words)
    
    return cleaned_text
```

```python
#drop updated_movie_name column if it exists
if 'updated_movie_name' in new_convoDF.columns:
    new_convoDF = new_convoDF.drop(columns=['updated_movie_name'])
```

```python
#clean the meta.movie_name column
new_convoDF['updated_movie_name'] = new_convoDF['meta.movie_name'].apply(cleanMovieNames)
```

```python
treeVisualization = {}

for year in tqdm((release_Years)):
    decade = year // 10 * 10
    if decade not in treeVisualization:
        treeVisualization[decade] = {}
    for index, row in new_convoDF.iterrows():
        if row['meta.release_year'] == year:
            for genre in row['meta.genre']:
                if genre not in treeVisualization[decade]:
                    treeVisualization[decade][genre] = []
                utterances = uttDF.loc[(uttDF['meta.movie_id'] == row['meta.movie_idx']) & (uttDF['updated_utt'] != ''), 'updated_utt'].tolist()        #get all the utterances for the movie_id and insert it to a list 
                treeVisualization[decade][genre].append({'movie_name': row['updated_movie_name'], 'utterances': utterances})
```

## Preprocessing for force directed

```python
pairData = corpus.speaking_pairs(speaker_ids_only=True)
```

```python
convoDF = corpus.get_conversations_dataframe()
```

```python
import ast
#reset index
convoDF.reset_index(inplace=True)
#convert meta.genre column to list
convoDF['meta.genre'] = convoDF['meta.genre'].apply(lambda x: ast.literal_eval(x))
convoDF.head()
```

```python
# Check if the character is already in the nodes list
def is_character_in_nodes(character_name, nodes_list):
    for node in nodes_list:
        if node["character"] == character_name:
            return True
    return False



output={}
output['nodes']=[]
output['links']=[]

for pair in tqdm(pairData):
    id1 , id2 = pair
    speaker1 = corpus.get_speaker(id1)
    speaker2 = corpus.get_speaker(id2)
    movie_id = speaker1.meta['movie_idx']
    gender = speaker1.meta['gender']
    generes = convoDF[convoDF['meta.movie_idx'] == movie_id]['meta.genre'].iloc[0]      #get the movie genre from the movie_idx column
    movie_name = convoDF[convoDF['meta.movie_idx'] == movie_id]['meta.movie_name'].iloc[0]      #get the movie name from the movie_idx column
    movieYear = convoDF[convoDF['meta.movie_idx'] == movie_id]['meta.release_year'].iloc[0]      #get the movie year from the movie_idx column
   #check if movieYear can be converted to int then change movieYear to int and then into a decade
    if movieYear.isnumeric():
        movieYear = int(movieYear)
    else:
        movieYear = movieYear[0:4]
        movieYear = int(movieYear)
   #change the movieyear to decade
    movieYear = movieYear // 10 * 10
    
    # Check for existence before appending
    if not is_character_in_nodes(speaker1.meta['character_name'], output['nodes']):
        temp = {
            "id": speaker1.meta['character_name'],
            "gender": gender,
            "movie": movie_name,
            "year": movieYear,
            "genres": generes
        }
        output['nodes'].append(temp)

    #set freq to be random between 1 and 26
    freq = np.random.randint(1, 26)
    
    temp2={
        "source":speaker1.meta['character_name'],
        "target":speaker2.meta['character_name'],
        "value":freq,
        "movie": movie_name
    }
    output['links'].append(temp2)
    

```

```python
#check if the nodes list has any duplicates
def has_duplicates(listOfNodes):
    for i in tqdm(listOfNodes):
        if listOfNodes.count(i) > 1:
            return True
    return False

has_duplicates(output['nodes'])

```
